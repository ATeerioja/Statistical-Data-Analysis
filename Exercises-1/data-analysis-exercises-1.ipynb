1.3 Thyroid Disease (continued)

```python
#First create a filter for t/f values
yes_no_values = {"f","t"}
#Iterate over all columns, and check if it contains yes or no values
yes_no_columns = [col for col in data1.columns
                  if set(data1[col].dropna().unique()).issubset(yes_no_values)
                 ]
#Select only the wanted columns from variable
yes_no_data1 = data1[yes_no_columns]
#Convert t values to 1 and f to 0
yes_no_data1 = yes_no_data1.map(lambda x: 1 if x == "t" else 0)
#Calculate ratio for yes values / total observations
yes_ratio = yes_no_data1.sum() / len(yes_no_data1)

yes_ratio
```

```python
#Create subset with wanted columns
subset1 = data1[["TSH","T3","TT4","T4U","FTI","TBG"]]
#Square elementwise and the sum divided by count of all non nan values
subset1_ratio = (subset1**2).sum() / subset1.count()
subset1_ratio
```

```python
#Isolate the wanted columns
subset2 = data1[["T3","TT4"]]
meanT3 = subset2["T3"].mean()
#Replace NaN values with the mean of the column
subset2.loc[:, "T3"] = subset2["T3"].fillna(meanT3)
subsetratio = (subset2["T3"] / subset2["TT4"]).sum() / (subset2["T3"] / subset2["TT4"]).count()
print(f"Mean ratio of T3 and TT4 {(subsetratio):.4f}")
```

1.4 Purchases

```python
purchases = pd.read_csv("purchases.csv", sep=",", na_values=["NA"])
#There are postal adresses in the location column
#Changing them to the city name
purchases["location"] = purchases["location"].replace({"33100": "Tampere","20100": "Turku"})
purchases["sex"] = purchases["sex"].replace({"nale": "male"})
purchases["purchases"] = purchases["purchases"].fillna(0)

purchases.loc[60:86]
```

```python
#Mean imputation for retention_time
purchases_retention_time_mean = purchases["retention_time"].mean()
purchases["retention_time"] = purchases["retention_time"].fillna(purchases_retention_time_mean)
```

Bonus

```python
#Single statement for grouping, then filling NaN numbers with median for each variable
purchases["retention_time"] = purchases.groupby(["sex","location"])["retention_time"].transform(lambda x: x.fillna(x.median()))
```
