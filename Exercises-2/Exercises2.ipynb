2. Cyclists

```python
#Imports here
import pandas as pd
from scipy import stats
```

```python
cyclistsEspoo = pd.read_csv('cyclists-espoo.csv')
cyclistsHelsinki = pd.read_csv('cyclists-helsinki.csv')

#Yhdistetään kaikki arvot, myös tyhjät yhteen listaan ensimmäisen rivin mukaan.
cyclistsAll = cyclistsEspoo.merge(cyclistsHelsinki, how='left', on="date")

cyclistsAll.tail()
```

```python
cyclistsAll.date = pd.to_datetime(cyclistsAll.date)
data1 = cyclistsAll.sort_values(by=['date'])
print(data1['date'].unique())
#Mittauksia suoritettiin 2714 eri päivänä
```

```python
data2 = cyclistsAll
print(data2.count())
#Tells how many not nan values are there for each column
```

```python
data2.notna().all(axis=1).sum()
#How many rows have none nan values counted
```

```python
data2.sum()
#It would seem that Merikannontie is the busiest road in terms of number of cyclists
```

```python
data2.isna().sum()
#This shows that many of the streets are missing a lot of values
#This could mean that the busies street is actually some other street
#This shows that we need to be careful when interpreting data
```

3. Human heights

```python
heightdataset1 = pd.Series([170, 192, 184, 168, 176, 181, 163])
heightdataset2 = pd.Series([170, 170, 170, 170, 192, 192, 192, 192, 184, 184, 184, 184, 168, 168, 168, 168, 176, 176, 176, 176, 181,
181, 181, 181, 163, 163, 163, 163])
```

```python
#Histogram and density plot of two datasets
heightdataset1.plot(kind='density')
heightdataset2.plot(kind='density')
```

```python
heightdataset1.plot(kind='hist',bins=6)
```

```python
heightdataset2.plot(kind='hist',bins=6)
```

Based on the plots, I would consider these two series to be normally distributed. This is based purely on visuals, so the result is not concrete.

```python
data2 = stats.normaltest(heightdataset2)
data2.statistic
#Test result for data2 shows that it is not normally distributed with a p-value of > 0.05
```
